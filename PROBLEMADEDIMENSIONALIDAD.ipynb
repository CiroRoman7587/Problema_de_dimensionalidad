```python

# EJEMPLO PROBLEMA DE DIMENSIONALIDAD

# La maldición de la dimensionalidad

# ¿Qué son las dimensiones?
#   En el contexto del análisis de datos y el aprendizaje automático,
#   las dimensiones hacen referencia a las características
#   o atributos de los datos. Por ejemplo, si
#   consideramos un conjunto de datos de casas,
#   las dimensiones podrían incluir el precio de la casa, el tamaño, la cantidad de habitaciones, la ubicación, etc.

# ¿Cómo se produce la maldición de la dimensionalidad?
#   A medida que añadimos más dimensiones a nuestro conjunto de datos, el volumen del
#   espacio aumenta exponencialmente. Esto significa que los datos se vuelven escasos.
#   Piénsalo de esta manera: si tienes una línea (1D), es fácil rellenarla con unos pocos puntos. Si
#   tienes un cuadrado (2D), necesitas más puntos para cubrir el área. Ahora, imagina un cubo
#   (3D): necesitarías incluso más puntos para rellenar el espacio. Este concepto se extiende a
#   dimensiones superiores, lo que hace que los datos sean extremadamente escasos.

# ¿Qué problemas ocasiona?

#   1-Escasez de datos: Como se mencionó, los datos se vuelven escasos,
#     lo que significa que la mayor parte del espacio de alta dimensión está vacío.
#     Esto dificulta las tareas de agrupamiento y clasificación.


#   2-Mayor capacidad computacional: Más dimensiones significan más recursos
#   computacionales y tiempo para procesar los datos.


#   3-Sobreajuste: Con dimensiones mayores, los modelos pueden volverse excesivamente
#   complejos y ajustarse al ruido en lugar de al patrón subyacente. Esto reduce la
#    capacidad del modelo para generalizarse a nuevos datos.


#    4-Las distancias pierden significado: En dimensiones altas, la diferencia de distancias
#     entre puntos de datos tiende a volverse insignificante, lo que hace que medidas como
#     la distancia euclidiana sean menos significativas.


#    5-Degradación del rendimiento: Los algoritmos, especialmente aquellos que se basan en
#     mediciones de distancia como k-nearest neighbors , pueden experimentar una caída en
#     el rendimiento.

#    6- Desafíos de visualización. Los datos de alta dimensión son difíciles de visualizar, lo que
#     dificulta el análisis exploratorio de datos.


#     ¿Por qué ocurre la maldición de la dimensionalidad?

#     Esto ocurre principalmente porque, a medida que agregamos más características o
#     dimensiones, aumentamos la complejidad de nuestros datos sin aumentar necesariamente
#     la cantidad de información útil. Además, en espacios de alta dimensión, la mayoría de los
#     puntos de datos se encuentran en los "bordes" o "esquinas", lo que hace que los datos sean
#     dispersos.


#     Cómo resolver la maldición de la dimensionalidad

#     La solución principal a la maldición de la dimensionalidad es la "reducción de la
#     dimensionalidad". Se trata de un proceso que reduce el número de variables aleatorias en
#     consideración mediante la obtención de un conjunto de variables principales. Al reducir la
#     dimensionalidad, podemos conservar la información más importante de los datos y
#     descartar las características redundantes o menos importantes.



#     ¿Que es el PCA y para que se utiliza en clustering?


#     El PCA es un método estadístico que transforma las variables originales en un nuevo
#     conjunto de variables, que son combinaciones lineales de las variables originales. Estas
#     nuevas variables se denominan componentes principales.

#     Supongamos que tenemos un conjunto de datos que contiene información sobre distintos
#     aspectos de los automóviles, como potencia, par motor, aceleración y velocidad máxima.
#     Queremos reducir la dimensionalidad de este conjunto de datos mediante PCA.


#     Con el análisis de componentes principales, podemos crear un nuevo conjunto de variables
#     llamadas componentes principales. El primer componente principal capturaría la mayor
#     varianza en los datos, que podría ser una combinación de potencia y par motor. El segundo
#     componente principal podría representar la aceleración y la velocidad máxima. Al reducir la
#     dimensionalidad de los datos con el análisis de componentes principales, podemos
#     visualizar y analizar el conjunto de datos de manera más eficaz.

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
 np.random.seed(0)
num_points = 100
```

```python
data_2d = np.random.rand(num_points, 2)

# 3D
data_3d = np.random.rand(num_points, 3)

# 10D
data_10d = np.random.rand(num_points, 10)
```

```python
# Función para calcular distancias
def pairwise_distances(data):
    from scipy.spatial.distance import pdist, squareform
    distances = pdist(data)
    return distances
```

```python
dist_2d = pairwise_distances(data_2d)
dist_3d = pairwise_distances(data_3d)
dist_10d = pairwise_distances(data_10d)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.scatter(data_2d[:, 0], data_2d[:, 1], alpha=0.6)
plt.title("Datos en 2D")
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
```

```python
ax = plt.subplot(132, projection='3d')
ax.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], alpha=0.6)
ax.set_title("Datos en 3D")
ax.set_xlabel("Dim 1")
ax.set_ylabel("Dim 2")
ax.set_zlabel("Dim 3")
```

```python
plt.subplot(133)
plt.hist(dist_10d, bins=30, alpha=0.7, color='r')
plt.title("Distancias en 10D")
plt.xlabel("Distancia")
plt.ylabel("Frecuencia")
```

```python
plt.tight_layout()
plt.show()
```
